```python
def one_hot(dataframe, columns):
    for col in columns:
        col = pd.get_dummies(dataframe[col], drop_first=True)
        dataframe = pd.concat([dataframe, col], axis=1)
        dataframe.drop(col, axis=1)
    return dataframe
```
